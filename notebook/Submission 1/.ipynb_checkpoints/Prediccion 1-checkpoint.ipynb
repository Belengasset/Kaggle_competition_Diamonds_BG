## Prediccion 1

```python
import pickle
import pandas as pd
```

```python
df_test = pd.read_csv("df_test.csv", sep=',', encoding='UTF-8')
df_test.head()
```

```python
# standardization
with open('estandarizacion.pkl', 'rb') as estandarizacion:
    estandarizacion = pickle.load(estandarizacion)
```

```python
# model
with open('mejor_modelo.pkl', 'rb') as modelo:
    modelo = pickle.load(modelo)
```

```python
# we standardize all the columns that are of numeric type using the transformer that we saved when we made the model

estandarizadas = pd.DataFrame(estandarizacion.transform(df_test[["carat", "depth", "table", 
                           "x", "y", "z"]]), columns = ["carat", "depth", "table", 
                           "x", "y", "z"])
estandarizadas
```

```python
# encoding cut
with open('data/encodingcut.pkl', 'rb') as cut:
    encoding_cut = pickle.load(cut)

# encoding color
with open('data/encodingcolor.pkl', 'rb') as color:
    encoding_color = pickle.load(color)

# encoding clarity
with open('data/encodingclarity.pkl', 'rb') as clarity:
    encoding_clarity = pickle.load(clarity)
```

```python
df_test = df_test.drop(['carat','depth', 'table', 'x', 'y', 'z'], axis =1)
df_test.head()
```

```python
df_final = pd.concat([estandarizadas, df_test], axis = 1)
df_final
```

```python
df_final = df_final.reindex(['id', 'carat','depth', 'table', 'x', 'y', 'z',
                             'cut_encoded', 'color_encoded', 'clarity_encoded',], axis=1)
df_final
```

```python
arr = modelo.predict(df_final)
```

```python
df = pd.DataFrame(arr, columns=['price'])
df.insert(0, 'id', range(0, len(df)))
df.head()
```

```python
nombre_archivo = 'submission1.csv'
```

```python
df_final.to_csv(nombre_archivo, index=False)
```
