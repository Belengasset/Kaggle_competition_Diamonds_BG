```python
import pickle
import pandas as pd
```

```python
df_test = pd.read_csv("df_test.csv", sep=',', encoding='UTF-8')
df_test.head()
```

```python
# estandarización
with open('estandarizacion.pkl', 'rb') as estandarizacion:
    estandarizacion = pickle.load(estandarizacion)
```

```python
# cargamos el mejor modelo
with open('mejor_modeloKNN.pkl', 'rb') as modelo:
    modelo = pickle.load(modelo)
```

```python
# estandarizamos todas las columnas que son de tipo numérico usando el transformer que nos guardamos cuando hicimos el modelo

estandarizadas = pd.DataFrame(estandarizacion.transform(df_test[["carat", "depth", "table", 
                           ]]), columns = ["carat", "depth", "table", 
                           ])
estandarizadas
```

```python
df_test = df_test.drop(['carat','depth', 'table'], axis =1)
df_test.head()
```

```python
df_final = pd.concat([estandarizadas, df_test], axis = 1)
df_final
```

```python
df_final = df_final.reindex(['id', 'carat','depth', 'table',
                             'color_oe', 'cut_oe', 'clarity_oe',], axis=1)
df_final
```

```python
# hacemos la predicción para nuestro nuevo paciente
arr = modelo.predict(df_final)
```

```python
df = pd.DataFrame(arr, columns=['price'])
df.insert(0, 'id', range(0, len(df)))
df.head()
```

```python
nombre_archivo = 'submission4KNN.csv'
```

```python
df.to_csv(nombre_archivo, index=False)
```
